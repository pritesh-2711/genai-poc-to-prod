```python
# Until now, the assistant only has the knowledge that was used to train those models.
# The original goal that we set uout to do was make a research paper assistant.
# So the llm will only have knowledge about the research papers until a specific point in time.
# What if I want to give it more knowledge about the research papers that were published after that point in time?
# I can use the retrieval augmented generation (RAG) approach to give the assistant access to the latest research papers.
```

```python
# I will not go into the typical definition of RAG; but instead start with what are the building blocks of RAG.
# I will add each block one by one and then show how to put them together to create a RAG pipeline.
# The foundational building blocks of RAG are:
# - Document extraction
# - Chunking
# - Embedding
# - Indexing into Vector database
# - Retrieval
# - Generation 
```

```python
# On top of the foundational building blocks, 
# we can also add some additional blocks to make the RAG pipeline more robust and efficient.

# Additional building blocks of RAG are:
# - Query Analysis
# - Re-ranking
# - Validation
# - Feedback loop
```

```python
# Tools, Agents and RAG are not mutually exclusive.
# We can use tools and agents in the RAG pipeline to make it even more efficient and robust; 
# but I'll pick it up after we have successfully implemented RAG pipeline without tools and agents.
```

```python
from langchain_community.document_loaders import TextLoader
```

```python
loader = TextLoader(file_path=data_directory+"HRA.txt")
```
