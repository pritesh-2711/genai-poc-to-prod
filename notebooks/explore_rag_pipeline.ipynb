```python
# PDFPlumber - Convert Tables to Structured Formats
# Once extracted, tables can be stored in multiple formats for RAG:

def extract_and_structure_tables(pdf_path):
    """
    Extract tables from PDF and convert to structured formats
    Handles duplicate column names by making them unique.
    Returns: dict with page_number, table_index, dataframe, csv_string, json_string
    """
    from collections import Counter
    
    structured_tables = []
    
    def make_unique_columns(columns):
        """Handle duplicate column names by appending suffixes"""
        counts = Counter(columns)
        result = []
        column_counters = Counter()
        
        for col in columns:
            if counts[col] > 1:
                column_counters[col] += 1
                result.append(f"{col}_{column_counters[col]}")
            else:
                result.append(col)
        return result
    
    with pdfplumber.open(pdf_path) as pdf:
        for page_num, page in enumerate(pdf.pages):
            tables = page.extract_tables()
            
            if not tables:
                continue
            
            for table_idx, table in enumerate(tables):
                if len(table) < 2:  # Skip malformed tables
                    continue
                
                # Create DataFrame with original columns
                df = pd.DataFrame(table[1:], columns=table[0])
                original_columns = list(df.columns)
                
                # Make columns unique if needed
                unique_columns = make_unique_columns(original_columns)
                df.columns = unique_columns
                
                try:
                    # Store in multiple formats
                    structured_tables.append({
                        'page': page_num + 1,
                        'table_index': table_idx,
                        'dataframe': df,
                        'csv': df.to_csv(index=False),
                        'json': df.to_json(orient='records'),
                        'markdown': df.to_markdown(index=False),
                        'columns': unique_columns,
                        'original_columns': original_columns,
                        'rows': len(df),
                        'metadata': {
                            'source': pdf_path.split('/')[-1],
                            'page_number': page_num + 1,
                            'table_id': f"page_{page_num + 1}_table_{table_idx}",
                            'has_duplicate_columns': len(original_columns) != len(set(original_columns))
                        }
                    })
                except Exception as e:
                    print(f"Error processing table on page {page_num + 1}, table {table_idx}: {str(e)}")
                    continue
    
    return structured_tables
```

```python
# Example usage
tables_data = extract_and_structure_tables(data_directory + filename)
```

```python
# Summary of extracted tables
print(f"\nTotal tables extracted: {len(tables_data)}\n")
print("=" * 80)

for i, table_info in enumerate(tables_data[:3]):  # Show first 3 tables
    metadata = table_info['metadata']
    print(f"\nTable {i+1}: {metadata['table_id']}")
    print(f"  Has duplicate columns: {metadata['has_duplicate_columns']}")
    print(f"  Shape: {table_info['rows']} rows x {len(table_info['columns'])} columns")
    print(f"  Columns: {table_info['columns']}")
    print(f"  Original columns: {table_info['original_columns']}")
    print("\n  Data sample:")
    print(table_info['dataframe'].head(2))
    print("-" * 80)
```

```python
# PRACTICAL USE CASES FOR EXTRACTED TABLES

# Use Case 1: Query answering with structured data
# Now you CAN answer questions like: "What was the average of total liabilities?"

# Example: Find numeric columns and compute aggregations
if tables_data:
    first_table = tables_data[0]
    df = first_table['dataframe']
    
    print("=" * 80)
    print(f"Table: {first_table['metadata']['table_id']}")
    print(f"Duplicate columns detected: {first_table['metadata']['has_duplicate_columns']}")
    print("=" * 80)
    
    # Find numeric columns
    numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
    
    if numeric_cols:
        print(f"\nNumeric columns found: {numeric_cols}")
        print("\nPossible aggregations:")
        for col in numeric_cols[:3]:  # Show first 3
            try:
                values = pd.to_numeric(df[col], errors='coerce').dropna()
                if len(values) > 0:
                    print(f"  {col}:")
                    print(f"    - Mean: {values.mean():.2f}")
                    print(f"    - Sum: {values.sum():.2f}")
                    print(f"    - Min: {values.min():.2f}")
                    print(f"    - Max: {values.max():.2f}")
            except Exception as e:
                print(f"    Error computing stats: {str(e)}")
    
    # Use Case 2: Export for RAG system
    print("\n" + "=" * 80)
    print("EXPORTING FOR RAG SYSTEM:")
    print("=" * 80)
    
    # As structured JSON
    print("\n1. JSON Format (for API/vector DB):")
    print(f"   Size: {len(first_table['json'])} bytes")
    print(f"   First 200 chars: {first_table['json'][:200]}...")
    
    # As CSV
    print("\n2. CSV Format (for spreadsheet integration):")
    csv_lines = first_table['csv'].split('\n')
    print(f"   Rows: {len(csv_lines) - 1}")  # -1 for header
    print(f"   Headers: {csv_lines[0]}")
    
    # As Markdown
    print("\n3. Markdown Format (for documentation):")
    md_lines = first_table['markdown'].split('\n')
    for line in md_lines[:5]:
        print(f"   {line}")
```

```python
import fitz  
from PIL import Image
import io
import hashlib
import numpy as np
from pathlib import Path

try:
    import cv2
except ImportError:
    print("OpenCV not installed, will use PIL for basic image analysis")
    cv2 = None
```

```python
# Image Detection with PDFPlumber
# PDFPlumber provides a quick way to detect images and their positions

pdf_file = data_directory + filename

with pdfplumber.open(pdf_file) as pdf:
    print(f"Total pages: {len(pdf.pages)}\n")
    
    image_count = 0
    for page_num, page in enumerate(pdf.pages):
        # Get all images on the page
        images = page.images
        
        if images:
            image_count += len(images)
            print(f"Page {page_num + 1}: Found {len(images)} image(s)")
            
            for img_idx, img in enumerate(images):
                # img is a dict with keys: top, bottom, left, right, width, height, srcsize
                print(f"  Image {img_idx + 1}:")
                print(f"    Position: ({img['x0']:.1f}, {img['top']:.1f})")
                print(f"    Size: {img['width']:.1f}x{img['height']:.1f}")
                print(f"    Source size: {img.get('srcsize', 'Unknown')}")

print(f"\nTotal images found: {image_count}")
```

```python
# Detailed Image Extraction with PyMuPDF
# PyMuPDF (fitz) gives access to image XObjects and their metadata

pdf_file_path = data_directory + filename
doc = fitz.open(pdf_file_path)

print(f"Total pages: {len(doc)}\n")

image_metadata = []

for page_num in range(len(doc)):
    page = doc[page_num]
    
    # Get all image XObjects on the page
    image_list = page.get_images()
    
    if image_list:
        print(f"Page {page_num + 1}: Found {len(image_list)} image(s)")
        
        for img_index, img_ref in enumerate(image_list):
            # img_ref is a tuple: (xref_number,)
            xref = img_ref[0]
            
            # Get image information
            pix = fitz.Pixmap(doc, xref)
            
            # Extract basic metadata
            img_info = {
                'page': page_num + 1,
                'image_index': img_index,
                'xref': xref,
                'width': pix.width,
                'height': pix.height,
                'size_pixels': pix.width * pix.height,
                'n': pix.n,  # Number of components (3=RGB, 4=RGBA, 1=Gray, etc.)
                'colorspace': pix.colorspace.name if pix.colorspace else None,
                'alpha': pix.alpha,  # Has alpha channel
            }
            
            # Get image bytes and calculate compressed size
            img_bytes = pix.tobytes()
            img_info['uncompressed_bytes'] = len(img_bytes)
            
            # Calculate image hash for deduplication
            img_info['hash'] = hashlib.md5(img_bytes).hexdigest()
            
            # Get aspect ratio
            img_info['aspect_ratio'] = pix.width / pix.height if pix.height > 0 else 0
            
            print(f"  Image {img_index + 1}:")
            print(f"    Dimensions: {img_info['width']}x{img_info['height']} px")
            print(f"    Colorspace: {img_info['colorspace']}")
            print(f"    Size: {img_info['uncompressed_bytes'] / 1024:.1f} KB")
            print(f"    Aspect ratio: {img_info['aspect_ratio']:.2f}")
            
            image_metadata.append(img_info)
            
            # Clean up
            pix = None

doc.close()
print(f"\nTotal images extracted: {len(image_metadata)}")
```
